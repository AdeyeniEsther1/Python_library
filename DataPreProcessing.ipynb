## STRATEGY TO HANDLE MISSING VALUES
The strategy to handle the missing values in this dataset is to drop missing values or fill the missing values with zero,in which I did.
The dropping of missing values will automotaically remove rows and colums containing missing values entirely.
Meanwhile, the approach I used with this dataset was filling the values with 0.In this case, all missing values will be replaced with 0.      
# Why I made this choice?
- Simplicity: Replacing missing values with zero is a straightforward and easy-to-implement method. It doesn't require complex calculations or advanced statistical techniques.
- Preserves Data Structure: Unlike dropping rows or columns, replacing with zero maintains the original structure of the DataFrame, which can be helpful for some analyses, especially if the number of missing values is small.
- Certain Scenarios: In specific cases, missing values might genuinely represent a zero value. For example, if a column tracks "number of purchases," a missing value could represent a customer who didn't buy anything (resulting in a zero value).

```python
#Transform the 'counter_type' feature to a numeric variable using the encoder of your choice
#import the labelEncoder from the scikit-learn library
from sklearn.preprocessing import LabelEncoder
#Create a labelEncoder object
encoder = LabelEncoder()
# Fit the encoder to the 'counter_type' feature
encoder.fit(df['counter_type'])
# Transform the 'counter_type' feature (categorical to numerical)
df['counter_type_encoded'] = encoder.transform(df['counter_type'])
```
